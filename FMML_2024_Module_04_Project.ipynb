### Gradient Descent

````python
##########################TASK 1
##########################

# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

def df(x):
    # [REQUIRED] Define the derivative (slope) of your function here
    return 2 * x - 4

def f(x):
    # [REQUIRED] Define your function here
    return x**2 - 4*x + 4

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)  # [REQUIRED] Calculate the gradient (slope) at current x
    new_x = previous_x - step_multiplier * gradient  # [REQUIRED] Update x using gradient descent

    step_size = abs(new_x - previous_x)
    # print(step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x))

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
````

## Gradient Descent

````python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
````

#### Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

#### $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

#### $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

#### MSE & Gradient Descent

````python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
````

````python

import numpy as np

# Define the Mean Squared Error function
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Define the gradient function for linear regression
def grad(x, y, thetas):
    m = len(y)
    predictions = thetas[0] + thetas[1] * x  # y = theta_0 + theta_1 * x
    error = predictions - y
    gradient_0 = (2/m) * np.sum(error)  # Gradient for Theta 0
    gradient_1 = (2/m) * np.sum(error * x)  # Gradient for Theta 1
    return np.array([gradient_0, gradient_1])

# Example data (ensure x_5 and y_5 are defined properly)
x_5 = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshape x_5 as a column vector
y_5 = np.array([1, 2, 2.8, 4.1, 5.2])  # Target (1D array)

# Initialize parameters (theta_0, theta_1)
thetas = np.array([2.9, 2.9])

# Learning rate
multiplier = 0.01

# Store the values for plotting (theta_0, theta_1)
plot_vals = thetas.reshape(1, 2)
mse_vals = mse(y_5, thetas[0] + thetas[1]*x_5.reshape(-1))  # Flatten x_5 to match predictions

# Gradient Descent
for i in range(1000):
    thetas = thetas - multiplier * grad(x_5, y_5, thetas)

    # Append the new values to our numpy arrays
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)

    # Calculate and append MSE for the current iteration
    mse_vals = np.append(mse_vals, mse(y_5, thetas[0] + thetas[1]*x_5.reshape(-1)))

# Results
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print('MSE is:', mse(y_5, thetas[0] + thetas[1]*x_5.reshape(-1)))
````
